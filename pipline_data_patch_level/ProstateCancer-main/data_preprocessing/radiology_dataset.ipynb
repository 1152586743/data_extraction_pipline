#Download PRAD radiology data
Since the public pathology image data is from TCGA-PRAD, we would also like to include radiology data into the study. Let's download and check the availability of this dataset.
The description of this radiology dataset can be found here.
https://www.cancerimagingarchive.net/collection/tcga-prad/ 

## Install the NBIA Data Retriever

``` bash
wget -P ./ https://github.com/CBIIT/NBIA-TCIA/releases/download/DR-4_4_3-TCIA-20240916-1/nbia-data-retriever_4.4.3-1_amd64.deb
dpkg -x package.deb ./
```
Then add the extracted folder to PATH
## Download the data
Download the Manifest file from this site, the file is ended with .tcia
https://www.cancerimagingarchive.net/collection/tcga-prad/

``` bash
cd /data/jjiang10/Data/PRAD_Radiology
```
click on search and download the manifest file, the file name in my case is "manifest-1739222542325.tcia"
Add your username and password to a txt file "credentials.txt"
``` text
userName=JJiang124
passWord=xxxxx
```
Use the command line tools to download the dataset
``` bash
nbia-data-retriever --cli './manifest-1739222542325.tcia' -d ./ -l ./credentials.txt --agree-to-license -v true
```
Check the size of the folder
``` bash
du -sh ./manifest-1739222542325
```
